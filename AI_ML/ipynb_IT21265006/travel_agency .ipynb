```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import joblib
from sklearn.metrics import accuracy_score

# Load the dataset
file_path = '/content/drive/MyDrive/weather/Contact Information (Responses) - Form Responses 1.xlsx'
data = pd.read_excel(file_path)

# Clean column names
data.columns = data.columns.str.strip()

# Define input features (X) and target column (Y)
input_features = [
    'Favorite Activities', 'Budget', 'Number of Persons', 'Transportation Options',
    'Month', 'Place', 'Health Condition', 'Accomadation Preferences',
    'Safty Concersn', 'Dieatary Needs', 'Pet Friendly Option', 'Connectivity',
    'Language Preference', 'Shoppping Preferences', 'Accessibility needs'
]
target_column = 'Package and Agency'

# Drop rows with missing target values
data = data.dropna(subset=[target_column])

# Encode categorical features and target
encoder = LabelEncoder()

for column in input_features + [target_column]:
    data[column] = encoder.fit_transform(data[column].astype(str))

# Split data into features (X) and target (Y)
X = data[input_features]
Y = data[target_column]

# Split into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train the Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, Y_train)

# Evaluate the model
Y_pred = model.predict(X_test)
print(classification_report(Y_test, Y_pred))

accuracy = accuracy_score(Y_test, Y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Save the trained model
joblib.dump(model, 'random_forest_model.pkl')
print("Model saved as 'random_forest_model.pkl'")
```
